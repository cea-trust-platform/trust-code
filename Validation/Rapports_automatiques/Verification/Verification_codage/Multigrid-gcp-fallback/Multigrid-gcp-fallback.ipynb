### Description
 
The multigrid solver must switch to the GCP solver if the former has not reached the required threshold. We previously stopped the computation if the threshold was not reached. This test case compare two pressure fields from two computations. The GCP fallback has been developed by Anida KHIZAR.

```python
from trustutils import run 
 
run.TRUST_parameters("1.9.3")
```

```python
run.reset()
run.initBuildDirectory()
run.addCase("THRESHOLD-1em10", "DNS_swarm.data", nbProcs=4)
run.addCase("THRESHOLD-1em12", "DNS_swarm.data", nbProcs=4)
run.printCases()
run.runCases()
```
